# Look at categorical features, their distributions and frequencies of groups within each category

## 1. flat type

```python

plt.figure(figsize=(8,4)) # this creates a figure 8 inch wide, 4 inch high

sns_plt = sns.countplot(df['flat_type'])

#sns_plt = sns.countplot(df['flat_model'])

sns_plt . set_title("Singapore: flat type")

#sns_plt . xlabel( fontsize=1)

#ax = sns.countplot(x="Column", data=ds)

sns_plt . set_xticklabels( sns_plt . get_xticklabels(), rotation=50, ha="right" )

x = sns_plt . set_xticklabels( sns_plt . get_xticklabels(), fontsize=10)

plt . show()
```

```python
# Frequency of each group within the feature "flat type"

plt.plot(df['flat_type'].value_counts().values)

plt.xlabel('flat type')

plt.ylabel(' Frequency')

plt . show()
```

- One can see that there are 7 flat types. Majority of flats have 4 or 3 rooms
